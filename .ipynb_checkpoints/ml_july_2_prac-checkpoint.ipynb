```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# make random data
np.random.seed(42)
x=2*np.random.rand(100,1)
y=4+3*x+np.random.randn(100,1)

# scale x and y
scaler_x=StandardScaler()
scaler_y=StandardScaler()
x_scaled=scaler_x.fit_transform(x)
y_scaled=scaler_y.fit_transform(y)

# build poly svr with solid params
svr_poly=SVR(kernel="poly",degree=3,C=100,epsilon=0.5)
svr_poly.fit(x_scaled,y_scaled.ravel())

# predict on smooth x for plotting
x_new=np.linspace(0,2,500).reshape(-1,1)
x_new_scaled=scaler_x.transform(x_new)
y_new_scaled=svr_poly.predict(x_new_scaled)
y_new=scaler_y.inverse_transform(y_new_scaled.reshape(-1,1))

# plotting
plt.figure(figsize=(12,6))
plt.plot(x,y,"b.",label="data points")
plt.plot(x_new,y_new,"r-",linewidth=2,label="svr prediction")
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16,rotation=0)
plt.title("svr with polynomial kernel",fontsize=18)
plt.legend(loc="upper left",fontsize=14)
plt.grid(True)

# plot epsilon margins
y_new_upper=scaler_y.inverse_transform((y_new_scaled+svr_poly.epsilon).reshape(-1,1))
y_new_lower=scaler_y.inverse_transform((y_new_scaled-svr_poly.epsilon).reshape(-1,1))
plt.plot(x_new,y_new_upper,"k--",label="epsilon margins")
plt.plot(x_new,y_new_lower,"k--")

plt.show()
```
